```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
%matplotlib inline
```

```python
x = np.array([1, 4, 8, 6, 3, 1, 4, 8])
y = np.array([2, 1, 1.5, 4, 5, 7, 8, 7.5])
```

```python
plt.axes().set_aspect(1)
plt.plot(x, y, 's')
```

```python
i = np.array([1, 2, 3, 4, 5, 6, 7, 8])
t = np.linspace(1, 8, 500)
```

```python
f = np.polyfit(i, x, 3)
g = np.polyfit(i, y, 3)
xx = np.polyval(f, t)
yy = np.polyval(g, t)
```

```python
plt.axes().set_aspect(1)
plt.plot(x, y, 's')
plt.plot(xx, yy)
```

```python
f = interpolate.interp1d(i, x, kind='cubic')
g = interpolate.interp1d(i, y, kind='cubic')
xx = f(t)
yy = g(t)
```

```python
plt.plot(x, y, 's')
plt.plot(xx, yy)
```

```python
d = np.array([])
for i in range(len(x)-1):
    dist = np.sqrt((x[i + 1] - x[i]) ** 2 + (y[i + 1] - y[i]) ** 2)
    d = np.append(d, np.array(dist))
d
```

```python
di = np.array([])
for j in range(len(x)):
    s = 0
    for i in range(j):
        s += d[i]
    di = np.append(di, s)
di
```

```python
t = np.linspace(0, 23, 500)
```

```python
f = np.polyfit(di, x, 3)
g = np.polyfit(di, y, 3)
xx = np.polyval(f, t)
yy = np.polyval(g, t)
```

```python
plt.axes().set_aspect(1)
plt.plot(x, y, 's')
plt.plot(xx, yy)
```

```python
f = interpolate.interp1d(di, x, kind='cubic')
g = interpolate.interp1d(di, y, kind='cubic')
xx = f(t)
yy = g(t)
```

```python
plt.plot(x, y, 's')
plt.plot(xx, yy)
```
